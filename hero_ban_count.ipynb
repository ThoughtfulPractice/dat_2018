```python
import pandas as pd

df = pd.read_csv('output/drafts.csv') #import drafts file 

heroes_df = pd.read_csv('output/basic_heroes.csv') #import heroes file
heroes_df = heroes_df [['id','localized_name']].set_index('id')
heroes_dict=heroes_df.to_dict() #create dctionary of heroes
heroes_dict = heroes_dict['localized_name']
```

```python
# Count number of times hero was banned
hero_bans = pd.DataFrame(df[df.pick == False]['hero_id'].value_counts()).reset_index()
hero_bans.columns=['hero_id','num_bans']

# Count number of times hero was picked
hero_picks = pd.DataFrame(df[df.pick == True]['hero_id'].value_counts()).reset_index()
hero_picks.columns=['hero_id','num_picks']
```

```python
# Merge the ban pick data and add column with hero name
merged=hero_bans.merge(hero_picks,how='outer',on='hero_id')
merged['hero_name']=merged['hero_id'].replace(heroes_dict) #replace hero_id with hero names
merged.fillna(0,inplace=True)
merged['sum_pickban']=merged['num_bans'].add(merged['num_picks'])
merged
```
